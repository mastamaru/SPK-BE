```python
clean_df = df.loc[:, ['Name', 'Genres', 'Studios']]
```

```python
clean_df = clean_df[clean_df['Studios'] != 'UNKNOWN']
clean_df = clean_df[clean_df['Genres'] != 'UNKNOWN']
```

```python
clean_df['Genres'] = clean_df['Genres'].apply(lambda x: x.split(', '))
```

```python
# iterate over the list of Genres column and create new column named base on the list values
for index, row in clean_df.iterrows():
    for genre in row['Genres']:
        clean_df.at[index, genre] = 1
```

```python
clean_df.fillna(0, inplace=True)
```
